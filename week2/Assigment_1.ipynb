1. Visualise the data you downloaded by placing a marker on a 2D plot for each
pair of feature values i.e. for each row in the data. On the plot the x-axis should
be the value of the first feature, the y-axis the value of the second feature and
the marker should be, for example, a + marker when the target value is +1 and
a o when the target is âˆ’1. Your plot should look similar in style to this (with
different data points of course!):

```python
import numpy as np
import pandas as pd
df = pd.read_csv("./data/week2.csv")
FeatureX = df.iloc[:, 0]
FeatureY = df.iloc[:, 1]
features = np.column_stack((FeatureX, FeatureY)) # Just create one 2D array
target = df.iloc[:, 2]
```

Now we have X we need to plot this data using matplotlib

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6)) # this plot graph size 
for i in range(len(target)):
    if target[i] == 1:
        plt.scatter(features[i][0], features[i][1], marker='+', color='blue', label='Class +1' if i == 0 else "")
    else:
        plt.scatter(features[i][0], features[i][1], marker='o', color='red', label='Class -1' if i == 0 else "")

plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.title('2D Feature Plot')
plt.legend()
```
